```python
marketing_cpy=marketing.copy()
```

```python
# clean up column names that contain whitespace

marketing_cpy.rename(columns={' Income ':'Income'}, inplace = True)
marketing_cpy["Income"]=marketing_cpy["Income"].str.replace("$","")


# transform Income column to a numerical

marketing_cpy["Income"]=marketing_cpy["Income"].str.replace(",","")
marketing_cpy["Income"]=marketing_cpy["Income"].astype(float)
```

```python
sns.boxplot(x='Income', data=marketing_cpy);
```

Changing the columns where there is an extra space and replacing  $ with a space and chaning the type of income column to float 

```python
a=marketing_cpy['Income'].median()
marketing_cpy['Income'].fillna(a,inplace=True)
```

Imputing the null values with a median as the distribution is skewed 

```python
marketing_cpy.head()
```

```python
sns.histplot(marketing_cpy['Income'],kde=20,bins=30);
```

```python
#transform Dt_Customer column to a Date time
marketing_cpy['Dt_Customer']=pd.to_datetime(marketing['Dt_Customer'])
```

Replace Birth year with age as it is more meaningful.As we would know the person belongs to which generation and what products he is interested to buy 

```python
marketing_cpy['Customer_Age_When_Enrolled'] = marketing_cpy['Dt_Customer'].dt.year - marketing_cpy['Year_Birth']
```

Combining some of the columns to a column by which it is useful for certain graphical analysis like

**Total Spent** which is 'MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds'
**Total_Purchases** which is 'NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth'

```python
Total_Spent=['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']
Total_Purchases=['NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth']
```

Combining all the ages of kids to see how children affect product buying

```python
marketing_cpy['Total_Spent']=(marketing_cpy['MntWines']+marketing_cpy['MntFruits']+marketing_cpy['MntMeatProducts']+marketing_cpy['MntFishProducts']+marketing_cpy['MntSweetProducts']+marketing_cpy['MntGoldProds'])
marketing_cpy['Total_Purchases']=(marketing_cpy['NumDealsPurchases']+marketing_cpy['NumWebPurchases']+marketing_cpy['NumCatalogPurchases']+marketing_cpy['NumStorePurchases']+marketing_cpy['NumWebVisitsMonth'])

# Merging columns Kids at home and Teenagers at home
marketing_cpy['Kids_Teen_at_home'] = marketing_cpy['Kidhome'] + marketing_cpy['Teenhome']
```

Droping the columns which are note useful like ID ,Kidhome,Teenhome

```python
# Reducing columns
marketing_cpy = marketing_cpy.drop(['ID','Kidhome','Teenhome'], axis = 1)
```

```python
sns.boxplot(x='Year_Birth', data=marketing_cpy);
```

As you can see from the box plot above ,there are outliers below 1900 where people are born are very less ,so we are not considering them

```python

marketing_cpy=marketing_cpy[marketing_cpy['Year_Birth']>=1920]
```

```python
marketing_cpy=marketing_cpy[marketing_cpy['Income']<150000]
```

There are outliers above income level of 10,0000 so we are not considering them 

```python
plt.grid();
plt.title('Income',size=15);
sns.histplot(marketing_cpy['Income'],kde=20)
```

This distribution looks normal after some cleaning and transformation

There are few labels in column marital _status like Alone ,YOLO ,Absurd which can be put under Single 

```python
marketing_cpy['Marital_Status'] = marketing_cpy['Marital_Status'].replace(['Alone','YOLO','Absurd'],'Single')
marketing_cpy.Marital_Status.value_counts()
```

We are combining the campaigns accepted into column TotalCampaignsAcc

```python
campaigns_cols = [col for col in marketing_cpy.columns if 'Cmp' in col] + ['Response'] # 'Response' is for the latest campaign
marketing_cpy['TotalCampaignsAcc'] = marketing_cpy[campaigns_cols].sum(axis=1)
```

```python
mnt_cols= [col for col in marketing_cpy.columns if 'Mnt' in col]
spending=pd.DataFrame(round(marketing_cpy[mnt_cols].mean(),1), columns=['Average']).sort_values(by='Average').reset_index()

#plot
ax=sns.barplot(x='Average', y='index',data=spending, palette='Blues')
plt.ylabel('Amount spent on...')

#add text labels for each bar's value
for p,q in zip(ax.patches, spending['Average']):
    ax.text(x=q+40,
           y=p.get_y()+0.5,
           s=q,
           ha='center');
```

```python
cam_success=pd.DataFrame(marketing_cpy[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']].mean()*100, columns=['Percent']).reset_index()

#plot
sns.barplot(x="Percent", y='index', data=cam_success.sort_values('Percent'),palette='Blues');
plt.xlabel('Accepted(%)');
plt.ylabel('Campaign');
plt.title('Marketing campaign success rate',size=16)
plt.show()
```
